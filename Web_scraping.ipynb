```python
titulo = []
autor = []
nota = []
n_aval = []
leram = []
lendo = []
querem_ler = []
relendo = []
abandonaram = []
resenhas = []
generos = []
favoritos = []

for livro in mais_lidos:
    url_livro = f'https://www.skoob.com.br/{livro}'
    response_livro = requests.get(url_livro)
    html_livro = response_livro.content
    soup_livro = BeautifulSoup(html_livro)
    
    titulo += [str(soup_livro.find_all('title')[0])[7:-8].split(" - ")[0]]
    autor += [str(soup_livro.find_all('title')[0])[7:-8].split(" - ")[1]]
    nota += [soup_livro.find_all(['star-rating'], attrs = {"size":"medium"})[0].get("rate")]
    n_aval += [soup_livro.find_all(['div'], attrs = {"id":"pg-livro-box-rating-avaliadores-numero"})[0].text.split(" ")[0]]
    leram += [soup_livro.find_all(['b'])[6].text]
    lendo += [soup_livro.find_all(['b'])[5].text]
    querem_ler += [soup_livro.find_all(['b'])[4].text]
    relendo += [soup_livro.find_all(['b'])[3].text]
    abandonaram += [soup_livro.find_all(['b'])[2].text]
    resenhas += [soup_livro.find_all(['b'])[1].text]
    generos += [soup_livro.find_all(['span'], attrs = {"class" : "pg-livro-generos"})[0].text]
    favoritos += [soup_livro.find_all(['div','a'], attrs = {"class":"link_11_off l_icones sk-cor-branco"})[0].text]
```
